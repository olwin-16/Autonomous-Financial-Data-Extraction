```python
import os
import time
import json
import logging
import requests
import gradio as gr
import torch
import asyncio
import faiss
import numpy as np
import re
from pathlib import Path
from typing import List, Dict, Any, Optional, Union
from PIL import Image
from bs4 import BeautifulSoup
from pydantic import BaseModel
from transformers import AutoProcessor, AutoModelForVision2Seq, AutoModelForCausalLM, Qwen2VLForConditionalGeneration
from sentence_transformers import SentenceTransformer, util
from playwright.async_api import async_playwright, Page
from playwright_stealth import stealth_sync
```

```python
# Configure Logging
LOGS_DIR = Path("Logs")
LOGS_DIR.mkdir(parents=True, exist_ok=True)

def log_to_file(message):
    with open(LOGS_DIR / "timing_logs.txt", "a") as f:
        f.write(message + "\n")

logging.basicConfig(
    level=logging.INFO,
    format="[%(asctime)s: %(levelname)s: %(message)s]",
    handlers=[logging.FileHandler(LOGS_DIR / "logs.log"), logging.StreamHandler()],
)
logger = logging.getLogger("UITarsLogger")
```

```python
# === UTILS FUNCTIONS ===
def encode_image(image_input: Union[Image.Image, bytes]) -> str:
    """Convert an image (PIL or bytes) to a base64-encoded string."""
    if isinstance(image_input, Image.Image):
        buffer = io.BytesIO()
        image_input.save(buffer, format="PNG")
        encoded_image = base64.b64encode(buffer.getvalue()).decode("utf-8")
    elif isinstance(image_input, bytes):
        encoded_image = base64.b64encode(image_input).decode("utf-8")
    else:
        raise ValueError(f"Expected Image or bytes, got {type(image_input)}")
    return encoded_image

def decode_base64_to_pil(encoded_image: str) -> Image.Image:
    """Convert a base64-encoded string back to a PIL Image."""
    image_data = base64.b64decode(encoded_image)
    return Image.open(io.BytesIO(image_data))
```

```python
# === LLM MODEL ===
class UITarsLLM:
    def __init__(self, model_name="UI-TARS-7B-DPO"):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.model = Qwen2VLForConditionalGeneration.from_pretrained(model_name, torch_dtype=torch.float16).to(self.device)
        self.processor = AutoProcessor.from_pretrained(model_name)

    def generate_response(self, user_input: str) -> str:
        inputs = self.processor(text=[user_input], return_tensors="pt").to(self.device)
        with torch.no_grad():
            generated_ids = self.model.generate(
                **inputs,
                max_new_tokens=150,
                do_sample=True,
                temperature=0.7,
                top_p=0.9,
                early_stopping=True,
                num_beams=1
            )
        return self.processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
# === VISION MODEL ===
class UITarsVision:
    def __init__(self, model_name="bytedance-research/UI-TARS-7B-DPO"):
        self.processor = AutoProcessor.from_pretrained(model_name)
        self.model = Qwen2VLForConditionalGeneration.from_pretrained(model_name)
        self.model.eval()

    def process(self, image: Image.Image, task: str = "Analyze GUI"):
        inputs = self.processor(images=image, text=task, return_tensors="pt")
        with torch.no_grad():
            outputs = self.model.generate(**inputs, max_length=256)
        decoded_output = self.processor.batch_decode(outputs, skip_special_tokens=True)
        return self.processor.batch_decode(outputs, skip_special_tokens=True)[0]
```

```python
async def enable_stealth(page):
    """Injects Stealth.js into the page to avoid bot detection."""
    with open("stealth.min.js", "r") as f:
        stealth_js = f.read()
    await page.add_script_tag(content=stealth_js)  # ✅ Inject JavaScript
```

```python
# === WEB SCRAPER ===
class LLMWebScraper:
    def __init__(self, headless: bool = True):
        self.headless = headless
        self.model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
        self.index = faiss.IndexFlatL2(384)
        self.embeddings = []
        self.chunks = []

    async def dynamic_search(self, query: str) -> List[str]:
      """Performs a Google search dynamically with stealth mode enabled."""
      async with async_playwright() as p:
          browser = await p.chromium.launch(headless=self.headless)  # Launch headless browser
          context = await browser.new_context()  # Use context for better stability
          page = await context.new_page()  # Create new page instance

          await enable_stealth(page)  # ✅ Properly injecting Stealth.js

          try:
              await page.goto("https://www.google.com/", timeout=60000)
              await page.wait_for_selector('input[name="q"]', timeout=30000)

              await page.fill('input[name="q"]', query)
              await page.press('input[name="q"]', "Enter")

              await page.wait_for_selector("#search", timeout=30000)

              links = await page.eval_on_selector_all(
                  "a[href]", "elements => elements.map(el => el.href)"
              )

          except Exception as e:
              print(f"🔴 Error during Google search: {e}")
              links = []

          await browser.close()
          return list(set(links[:5]))
```

```python
# === TOOLS FOR INTERACTIONS ===
async def click(state: Dict[str, Any], selector: str = None, position: tuple = None) -> str:
    page: Page = state["page"]
    try:
        if selector:
            await page.click(selector)
            return f"Clicked element: {selector}"
        elif position:
            await page.mouse.click(*position)
            return f"Clicked at position {position}"
        else:
            return "Failed to click: No selector or position provided."
    except Exception as e:
        return f"Failed to click: {str(e)}"

async def scroll(state: Dict[str, Any], direction: str = "down") -> str:
    page: Page = state["page"]
    try:
        scroll_amount = 500 if direction == "down" else -500
        await page.evaluate(f"window.scrollBy(0, {scroll_amount})")
        return f"Scrolled {direction}"
    except Exception as e:
        return f"Failed to scroll: {str(e)}"
```

```python
# === NAVIGATION TOOLS ===
async def navigate(state: Dict[str, Any], url: str) -> str:
    """Navigate to a specified URL."""
    page: Page = state["page"]
    try:
        await page.goto(url)
        return f"Navigated to {url}"
    except Exception as e:
        return f"Failed to navigate: {str(e)}"

async def go_back(state: Dict[str, Any]) -> str:
    """Navigates back to the previous page."""
    page: Page = state["page"]
    try:
        await page.go_back()
        return "Navigated back"
    except Exception as e:
        return f"Failed to go back: {str(e)}"
```

```python
# === MAIN GRADIO FUNCTION ===
async def fetch_website_info(query: str) -> str:
    scraper = LLMWebScraper(headless=True)
    result = await scraper.dynamic_search(query)
    log_to_file(f"Extracted Data: {result[:200]}...")
    log_to_file(f"Sources: {', '.join(result)}")
    return f"Sources: {', '.join(result)}\nQuestion: {query}\nAnswer:"

iface = gr.Interface(
    fn=lambda query: asyncio.run(fetch_website_info(query)),
    inputs=[gr.Textbox(label="Enter Your Question")],
    outputs=[gr.Textbox(label="Generated Response")],
    title="UI-TARS Web Agent",
    description="Ask a question, and the AI will dynamically scrape web content, analyze UI elements, and generate an answer."
)

iface.launch(share=True, debug=True)
```
